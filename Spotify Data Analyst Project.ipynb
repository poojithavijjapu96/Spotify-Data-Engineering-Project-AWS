```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
#spotipy is a module/package ; spotipy.oauth2 : a sub module ; SpotifyClientCredentials :  a class
```

```python
import pandas as pd
```

```python
client_credentials_manager = SpotifyClientCredentials(client_id="f9ece6168dc34bf8a855b6d9cbd9fe50", client_secret="517fb92932f548878076b0fb4fcb2e43")

# client_credentials_manager is an object for the class SpotifyClientCredentials with initian variables client_id and client_secret 
```

```python
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)

# sp object of class Spotify which is in spotipy module, and we are passing client_credentials_manager object for authorization
```

```python
playlist_link = "https://open.spotify.com/playlist/37i9dQZEVXbNG2KDcFcKOF?si=1333723a6eff4b7f"
```

```python
playlist_URI = playlist_link.split("/")[-1].split('?')[0]
# we are extracting the playlist id, like 37i9dQZEVXbNG2KDcFcKOF above
```

```python
data=sp.playlist_tracks(playlist_URI)

#extracting tracks using playlist_tracks function in sp object; the data is in json format
```

```python
data['items'][0]['track']['album']['id']
data['items'][0]['track']['album']['name']
data['items'][0]['track']['album']['release_date']
data['items'][0]['track']['album']['total_tracks']
data['items'][0]['track']['album']['external_urls']['spotify']
```

```python
album_list=[]
for row in data['items']:
    album_id=row['track']['album']['id']
    album_name=row['track']['album']['name']
    album_release_date=row['track']['album']['release_date']
    album_total_tracks=row['track']['album']['total_tracks']
    album_url=row['track']['album']['external_urls']['spotify']
    album_element={'album_id':album_id,'album_name':album_name, 'album_release_date':album_release_date,'album_total_tracks':album_total_tracks,'album_url':album_url}
    album_list.append(album_element)
```

```python
album_list
```

```python
data['items'][0]['track']['artists']
```

```python
artist_list=[]
for row in data['items']:
    for key,value in row.items():
        if key=='track':
            for artist in value['artists']:
                artist_dict={'artist_id':artist['id'],'artist_name':artist['name'],'external_uri':artist['href']}
                artist_list.append(artist_dict)
```

```python
artist_list
```

```python
song_list = []
for row in data['items']:
    song_id = row['track']['id']
    song_name = row['track']['name']
    song_duration = row['track']['duration_ms']
    song_url = row['track']['external_urls']['spotify']
    song_popularity = row['track']['popularity']
    song_added = row['added_at']
    album_id = row['track']['album']['id']
    artist_id = row['track']['album']['artists'][0]['id']
    song_element = {'song_id':song_id,'song_name':song_name,'duration_ms':song_duration,'url':song_url,
                    'popularity':song_popularity,'song_added':song_added,'album_id':album_id,
                    'artist_id':artist_id
                   }
    song_list.append(song_element)
```

```python
song_list
```

```python
album_df=pd.DataFrame.from_dict(album_list)
```

```python
album_df.info()
```

```python
album_df=album_df.drop_duplicates(subset='album_id')
```
